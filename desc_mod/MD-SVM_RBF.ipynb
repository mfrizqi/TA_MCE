```python
### LIBRARIES ###
import pandas as pd 
import numpy as np
import pickle
from tqdm import tqdm
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')
```

```python
train = pickle.load(open('./pickle/dtTrain.pkl', 'rb'))
test = pd.read_csv("data/TestSetNoId.csv")
```

```python
bd8 = pickle.load(open('./pickle/bestDesc8.pkl', 'rb'))
bd9 = pickle.load(open('./pickle/bestDesc9.pkl', 'rb'))
bd10 = pickle.load(open('./pickle/bestDesc10.pkl', 'rb'))
```

```python
train_8 = train.loc[:,bd8]
test_8 = test.loc[:,bd8]

train_9 = train.loc[:,bd9]
test_9 = test.loc[:,bd9]

train_10 = train.loc[:,bd10]
test_10 = test.loc[:,bd10]
```

```python
# 0. Preparation Data
x_trains = []
x_tests = []

x_trains.append(train_8.iloc[:,:])
x_trains.append(train_9.iloc[:,:])
x_trains.append(train_10.iloc[:,:])

x_tests.append(test_8.iloc[:,:])
x_tests.append(test_9.iloc[:,:])
x_tests.append(test_10.iloc[:,:])

y_train = train.iloc[:, [-1]]
y_test = test.iloc[:,[-1]]

x_trains[0].shape,x_tests[0].shape
```

```python
for i in tqdm(range(3)):
    #Initialize parameter grid
    param_grid={
            'C': [0.1, 1, 10, 100, 1000],
            'gamma': ['auto','scale'],
    }
    
    #create GridSearchCV object
    gsc = GridSearchCV(
        estimator=SVR(kernel='rbf'),
        param_grid=param_grid,
        cv=5, scoring='neg_mean_squared_error', verbose=0, n_jobs=-1)
    
    #Fit model
    gsc.fit(x_trains[i], y_train)
    
    # View the accuracy score
    print('Best score:', gsc.best_score_) 
    print('Best Param:',gsc.best_params_)
    
    model = SVR(C=gsc.best_params_['C'],kernel="rbf", gamma=gsc.best_params_['gamma'])
    model.fit(x_trains[i], y_train)
    
    y_train_pred = model.predict(x_trains[i])
    y_test_pred = model.predict(x_tests[i])

    r2_train = r2_score(y_train,y_train_pred)
    r2_test = r2_score(y_test,y_test_pred)

    print(r2_train, r2_test)
    print("===================")
```
